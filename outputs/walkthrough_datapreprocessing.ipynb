**Data pre-processing**
====================

In this walkthrough, I show the the processing I do as 
well the outputs I get. This is just to ensure that everyone 
is on the same page and also transparency. The data processing
step has the following steps:

* Create base event attributes
* Handle duplicate events
* Drop some events (test, duplicate, events before data collection)

1. ** Event attributes **

An event is simply an sms with. The following base attributes 
are recorded for each events:
* box_id: device id generated from message body
* datetime_sent: exact timestamp of event
* type_of_event (*pback, pfail, pfail_mon, pon_mon, test*)

The whole list of attributes is shown in figure below.

```python
Image(filename='event_attributes.png')
```

2 . ** Handling duplicates **

Based on the stata do file, I define duplicates as follows:
 * box_id, message_body & datetime_sent
 * box_id, datetime_sent

3. ** Dropping events**

The following type of events are dropped:
* Duplicate events (maintain only first one)
* Test event
* Events recorded before official start of data collection.
